```python
import os
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv("file1.csv" )
df1
```

```python
### creating a new column for the IATA Code, and accordingly editing the original column
df1['Destination_IATA'] = df1['Destination'].str[:3]
df1['Destination'] = df1['Destination'].str[4:]
df1
```

```python
df2=pd.read_csv("df2.csv")
```

```python
df2 = df2.rename(columns={'origin_IATA': 'Origin_IATA','destination_IATA': 'Destination_IATA'})
df2
```

```python
new_df = pd.merge(df1, df2, on=['Origin_IATA', 'Destination_IATA'])
new_df
```

```python
new_df = new_df.drop(['Operating airlines', 'Flight days', 'flight_duration'], axis=1)
```

```python
new_column_order = ['Origin', 'Origin_IATA','Flight_duration', 'Destination','Destination_IATA' , 'Flights_per_day',  'airline', 'distance (km)', 'aircraft type']
new_df = new_df.reindex(columns=new_column_order)
```

```python
new_df
```

```python
### auxiliary functions for data cleaning 
### will help us to prepare the data of 'aircraft type', 'Flights_per_day' and 'Flight_duration' columns 

def convert_flight_duration(duration):
    hours, minutes = duration.split('h ')
    return int(hours) * 60 + int(minutes[:-1])

def clean_flights_per_day(flights):
    if pd.isna(flights):
        return 0
    
    if isinstance(flights, str):
        flights = flights.replace(' flights per day', '')  # Remove 'flights per day' text
        
        if '-' in flights:
            # Split the range and calculate the average
            flights_range = flights.split('-')
            flights_min = flights_range[0].strip()
            flights_max = flights_range[1].strip()

            # Check if the range values are numeric
            if flights_min.isdigit() and flights_max.isdigit():
                flights_avg = (int(flights_min) + int(flights_max)) / 2
                return int(flights_avg)
            else:
                return 0
        else:
            # Extract the numeric part
            flights_val = flights.strip().split()[0]

            # Check if the value is numeric
            if flights_val.isdigit():
                return int(flights_val)
            else:
                return 0
    else:
        return int(flights)


def clean_aircraft_type(aircraft):
    if isinstance(aircraft, str):
        # Remove leading/trailing whitespace and split by space
        aircraft_list = aircraft.strip().split()

        # Remove any non-alphanumeric characters from each element
        aircraft_list = ["".join(c for c in a if c.isalnum()) for a in aircraft_list]

        # Remove empty elements
        aircraft_list = [a for a in aircraft_list if a]

        return aircraft_list
    else:
        return np.nan    
```

```python
copy_df=new_df.copy()
```

```python
copy_df['Operating airlines'] = copy_df.groupby(['Origin_IATA', 'Destination_IATA'])['airline'].transform('nunique')
copy_df
```

```python
copy_df.drop_duplicates(subset=['Origin_IATA', 'Destination_IATA'], keep='first', inplace=True)
copy_df.reset_index()
```

```python
### Make sure the number of NaN values is reasonable
for col in copy_df.columns:
    print(f"null in '{col}' -", copy_df[col].isnull().sum())
```

```python
copy_df.dropna(subset=['Flights_per_day'])
```

```python
copy_df['distance (km)'] = pd.to_numeric(copy_df['distance (km)'], errors='coerce')
copy_df['Flights_per_day'] = copy_df['Flights_per_day'].apply(clean_flights_per_day)
copy_df.dropna(subset=['Flights_per_day'])
copy_df['Flights_per_day'] = copy_df['Flights_per_day'].astype(int)
copy_df['Flight_duration'] = copy_df['Flight_duration'].apply(convert_flight_duration)
```

```python
copy_df.drop('airline',axis=1, inplace=True)
copy_df = copy_df.reset_index(drop=True)
```

```python
copy_df
```

```python
copy_df['aircraft type'] = copy_df['aircraft type'].apply(clean_aircraft_type)
copy_df
```

```python
### save the clean dataFrame
copy_df.to_csv('clean dataframe.csv', index=False)
```
